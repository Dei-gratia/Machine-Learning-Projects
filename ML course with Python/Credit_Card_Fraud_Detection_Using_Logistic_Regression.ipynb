```python
#import packages
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
#load dataset
credit_card_data = pd.read_csv("/content/creditcard.csv")
credit_card_data.head()
```

```python
credit_card_data.shape
```

```python
credit_card_data.info()
```

```python
credit_card_data["Class"].value_counts()
```

```python
legit = credit_card_data[credit_card_data.Class==0]
fraud = credit_card_data[credit_card_data.Class==1]
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
credit_card_data.groupby("Class").mean()
```

```python
#Under sampling
legit_sample = legit.sample(n=492)
new_credit_card_data = pd.concat([legit_sample, fraud], axis = 0)
new_credit_card_data.head()
```

```python
new_credit_card_data["Class"].value_counts()
```

```python
X = new_credit_card_data.drop(columns="Class", axis= 1)
Y = new_credit_card_data["Class"]
print(X)
print(Y)
```

```python
#splitting the dataset
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
print(X.shape, x_train.shape, x_test.shape)
```

```python
#training the model
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
#check model accuracy on training data
training_data_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(training_data_prediction, y_train)
print("Accuracy on training data: ", training_data_accuracy)
```

```python
#check model accuracy on test data
test_data_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(test_data_prediction, y_test)
print("Accuracy on test data: ", test_data_accuracy)
```
