```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
```

```python
gold_data = pd.read_csv("/content/gld_price_data.csv")
gold_data.head()
```

```python
gold_data.shape
```

```python
gold_data.info()
```

```python
gold_data.describe()
```

```python
correlation = gold_data.corr()
plt.figure(figsize=(8, 8))
sn.heatmap(correlation, cbar=True, square=True, fmt=".1f", annot=True, annot_kws={"size": 8}, cmap="Blues")
```

```python
print(correlation["GLD"])
```

```python
sn.distplot(gold_data["GLD"], color="green")
```

```python
X = gold_data.drop(["Date", "GLD"], axis=1)
Y = gold_data["GLD"]
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
print(X.shape, X_train.shape, X_test.shape)
```

```python
regressor_model = RandomForestRegressor()
regressor_model.fit(X_train, Y_train)
```

```python
train_data_prediction = regressor_model.predict(X_train)
error_score = metrics.r2_score(Y_train, train_data_prediction)
print("Training data R squared error: ", error_score)
```

```python
Y_train =list(Y_train)
plt.plot(Y_train, color="blue", label="Actual Value")
plt.plot(train_data_prediction, color="green", label="Predicted Value")
plt.title("Actual Price vs Predicted Price")
plt.xlabel("Number of values")
plt.ylabel("GLD Price")
plt.legend()
plt.show()
```

```python
test_data_prediction = regressor_model.predict(X_test)
error_score = metrics.r2_score(Y_test, test_data_prediction)
print("R squared error: ", error_score)
```

```python
Y_test =list(Y_test)
plt.plot(Y_test, color="blue", label="Actual Value")
plt.plot(test_data_prediction, color="green", label="Predicted Value")
plt.title("Actual Price vs Predicted Price")
plt.xlabel("Number of values")
plt.ylabel("GLD Price")
plt.legend()
plt.show()
```
