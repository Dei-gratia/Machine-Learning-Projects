```python
#import the libraries
import numpy as np
import pandas as pd
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
#creating the Lasso Regression
class Lassor_Regression():
  #initialize the hyperparameters
  def __init__(self, learning_rate, no_of_iterations, lambda_param) -> None:
    self.learning_rate = learning_rate
    self.no_of_iterations = no_of_iterations
    self.lambda_param = lambda_param



  #fit the dataset to the model
  def fit(self, x, y):
    self.x = x
    self.y = y
    #initialise number of features and datapoints
    self.m, self.n = x.shape
    #initialize the bias and weights
    self.w = np.zeros(self.n)
    self.b = 0
    #implement gradient descent
    for i in range(self.no_of_iterations):
      self.update_weights()



  #update model weights and bias
  def update_weights(self):
    y_pred = self.predict(self.x)

    #gradient for weights
    dw = np.zeros(self.n)
    for i in range(self.n):
      if self.w[i] > 0:
        dw[i] = (-(2*(self.x[:,i]).dot(self.y -y_pred)) + self.lambda_param) / self.m
      else:
        dw[i] = (-(2*(self.x[:,i]).dot(self.y -y_pred)) - self.lambda_param) / self.m

    #gradient for bias
    db = -2 * np.sum(self.y - y_pred) / self.m

    #updating the weights and bias
    self.w = self.w - self.learning_rate*dw
    self.b = self.b - self.learning_rate*db


  #predict the target value
  def predict(self, x):
    print(x)
    print(self.w)
    print(self.b)
    return x.dot(self.w) + self.b
```
