```python
#convert RGB images to grayscale image using OpenCV
img = cv2.imread("/content/cls.jpg")
type(img)
```

```python
grayscale_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
type(grayscale_img)
```

```python
grayscale_img.shape
```

```python
cv2_imshow(grayscale_img)
```

```python
# saving the grayscale image
cv2.imwrite("cls_grey.jpg", grayscale_img)
```
