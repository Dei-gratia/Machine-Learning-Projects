```python
#import dependances
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
from sklearn.preprocessing import StandardScaler
```

```python
#load the dataset
breast_cancer_dataset = load_breast_cancer()
print(breast_cancer_dataset)
```

```python
#load dataset to pd dataframe
breast_cancer_data = pd.DataFrame(breast_cancer_dataset.data, columns=breast_cancer_dataset.feature_names)
breast_cancer_data.head()
```

```python
breast_cancer_data["label"] = breast_cancer_dataset.target
breast_cancer_data.tail()
```

```python
#separate the features and targets
X = breast_cancer_data.drop(columns="label", axis=1)
y = breast_cancer_data["label"]
print(X)
print(y)
```

```python
#split the training and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
print(X.shape, X_train.shape, X_test.shape)
```

```python
#standadize the data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
print(X_train)
```

```python
##building the Neural network

#setting the layers of Neural Network
model = keras.Sequential([keras.layers.Flatten(input_shape=(30,)),
                          keras.layers.Dense(20, activation="relu"),
                          keras.layers.Dense(2, activation="sigmoid")])


#compile the Neural Network
model.compile(optimizer="adam",
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"])
```

```python
#train the Neural Network
history = model.fit(X_train, y_train, validation_split=0.1, epochs=10)
```

```python
#visualize the accuracy and loss
plt.plot(history.history["accuracy"])
plt.plot(history.history["val_accuracy"])

plt.title("model accuracy")
plt.ylabel("accuracy")
plt.xlabel("epoch")

plt.legend(["training data", "validation data"], loc = "lower right")
```

```python
plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])

plt.title("model loss")
plt.ylabel("loss")
plt.xlabel("epoch")

plt.legend(["training data", "validation data"], loc = "upper right")
```

```python
#check model accuracy on test data
loss, accuracy = model.evaluate(X_test, y_test)
print(accuracy)
```

```python
y_pred = model.predict(X_test)
print(y_pred)
```

```python
#convert the prediction probability to class labels
y_pred_labels = [np.argmax(i) for i in y_pred]
print(y_pred_labels)
```
