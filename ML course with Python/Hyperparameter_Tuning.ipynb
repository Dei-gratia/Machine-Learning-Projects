```python
#import libraries and models
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

```python
#load dataset
breast_cancer_dataset = load_breast_cancer()
print(breast_cancer_dataset)
```

```python
#convert the dataset into a datafram
breast_cancer_data = pd.DataFrame(breast_cancer_dataset.data, columns=breast_cancer_dataset.feature_names)
breast_cancer_data.head()
```

```python
#add the target column to the dataframe
breast_cancer_data['label'] = breast_cancer_dataset.target
breast_cancer_data.head()
```

```python
breast_cancer_data.shape
```

```python
breast_cancer_data.isnull().sum()
```

```python
#checking the distribution of target variables
breast_cancer_data['label'].value_counts()
```

```python
#separating the features and target
X = breast_cancer_data.drop(columns="label", axis=1)
y = breast_cancer_data["label"]
print(X)
print(y)
```

```python
X = np.asarray(X)
y = np.asarray(y)
```

```python
#load the svc model
model = SVC()

#hyperparametes
parameters = {
    "kernel":["linear", "poly", "rbf", "sigmoid"],
    "C":[1, 5, 10, 20]
}

#grid search
classifier = GridSearchCV(model, parameters, cv=5)

#fitting the data to the model
classifier.fit(X, y)
```

```python
classifier.cv_results_
```

```python
#check the best parameters
best_parameters = classifier.best_params_
print(best_parameters)
```

```python
#check the highest accuracy
highest_accuracy = classifier.best_score_
print(highest_accuracy)
```

```python
#load the results to pd dataframe
result = pd.DataFrame(classifier.cv_results_)
result.head()
```

```python
grid_search_result = result[['param_C', 'param_kernel', 'mean_test_score']]
grid_search_result
```

```python
#Randomized Search CV
classifier = RandomizedSearchCV(model, parameters, cv=5)

#fit the data to the model
classifier.fit(X, y)
```

```python
classifier.cv_results_
```

```python
#check the best parameters
best_parameters = classifier.best_params_
print(best_parameters)
```

```python
#check the highest accuracy
highest_accuracy = classifier.best_score_
print(highest_accuracy)
```

```python
#load the results to pd dataframe
result = pd.DataFrame(classifier.cv_results_)
result.head()
```

```python
rand_grid_search_result = result[['param_C', 'param_kernel', 'mean_test_score']]
rand_grid_search_result
```
