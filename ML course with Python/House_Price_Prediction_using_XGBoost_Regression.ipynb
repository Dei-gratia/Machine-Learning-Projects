```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn import metrics
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
#house_price_dataset = sklearn.datasets.load_boston()
housing = fetch_california_housing()
print(housing)
```

```python
housing_dataframe = pd.DataFrame(housing.data, columns=housing.feature_names)
housing_dataframe.head()
```

```python
housing_dataframe["price"] = housing.target
housing_dataframe.head()
```

```python
housing_dataframe.shape
```

```python
housing_dataframe.isnull().sum()
```

```python
housing_dataframe.describe()
```

```python
correlation = housing_dataframe.corr()
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```

```python
X = housing_dataframe.drop(["price"], axis=1)
Y = housing_dataframe["price"]
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = XGBRegressor()
model.fit(X_train, Y_train)
```

```python
training_data_prediction = model.predict(X_train)
print(training_data_prediction)
```

```python
score_1 = metrics.mean_absolute_error(Y_train, training_data_prediction)
score_2 = metrics.r2_score(Y_train, training_data_prediction)

print("Mean Absolute Error: ", score_1)
print("R Squared Error: ", score_2)
```

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Actual Price vs Predicted Price")
plt.show()
```

```python
test_data_prediction = model.predict(X_test)
score_1 = metrics.mean_absolute_error(Y_test, test_data_prediction)
score_2 = metrics.r2_score(Y_test, test_data_prediction)

print("Mean Absolute Error: ", score_1)
print("R Squared Error: ", score_2)
```
